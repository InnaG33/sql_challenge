```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
username= "postgres"
password= "XXX"
from sqlalchemy import create_engine
engine = create_engine(f'postgresql://{username}:{password}@localhost:5433/employeeshp')
connection = engine.connect()
```

```python
employees1 = pd.read_sql('SELECT * FROM "Employees"', connection)
employees1.head()
```

```python
employees1.to_csv("output/employees_read.csv")
```

```python
employees = pd.read_csv("output/employees_read.csv")
employees.head()
```

```python
salaries1 = pd.read_sql('SELECT * FROM "Salaries"', connection)
salaries1.head()
```

```python
salaries1.to_csv("output/salaries_read.csv")
```

```python
salaries = pd.read_csv("output/salaries_read.csv")
salaries.head()
```

```python
emp_salaries = pd.merge (salaries, employees, on=['emp_no'])
emp_salaries.head()
```

```python
grouped_salaries = emp_salaries.groupby(['emp_title_id'])
grouped_salaries.head(10)
```

```python
avg_salaries = grouped_salaries['salary'].mean()
avg_salaries
```

```python
from matplotlib import pyplot as plt
```

```python
# Create a histogram to visualize the most common salary ranges for employees

plt.hist(avg_salaries, 10, density=True, color='g', alpha=0.5)
plt.title("Common salaries by emplyer title_id", fontsize=16)
plt.xlabel("Avg Salaries", fontsize=14)
```

```python
# Create a bar chart of average salary by title

avg_salaries.plot.bar(rot=90, color='g', alpha=0.75, figsize=(12,5))
plt.title("Average Salaries per title IDs", fontsize=16)
plt.xlabel("Employer title IDs", fontsize=14)
plt.ylabel("Avg Salaries", fontsize=14)

```
